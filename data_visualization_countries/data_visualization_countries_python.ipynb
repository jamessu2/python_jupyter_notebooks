```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Basics of plotting with matplotlib

x = [1,2,3]
y = [1,4,9]
z = [10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y","this is z"])
plt.show()
```

```python
# Basics of accessing csv data with pandas

sample_data = pd.read_csv("sample_data.csv")
print(type(sample_data))
print(sample_data)
print("\n")

print(type(sample_data.column_c))
print(sample_data.column_c)
print("\n")

i = 1
print(type(sample_data.column_c.iloc[1]))
print(f"Value at index {i} = '" + str(sample_data.column_c.iloc[i]) + "'")
print("\n")
```

```python
# Basic of plotting csv data that's been access with pandas

plt.plot(sample_data.column_a, sample_data.column_b, 'o')
plt.plot(sample_data.column_a, sample_data.column_c)
plt.title("plot with data from sample_data.csv")
plt.xlabel("a")
plt.ylabel("b and c")
plt.legend(["b","c"])
plt.show()
```

```python
# Accessing and plotting a more complicated set of data

countries = pd.read_csv("countries.csv")
us_data = countries[countries.country == "United States"]
china_data = countries[countries.country == "China"]
```

```python
## Challenge #1: Compare the population growth of USA vs China

plt.plot(us_data.year, us_data.population / 10**6)
plt.plot(china_data.year, china_data.population / 10**6)
plt.title("population growth by country")
plt.xlabel("year")
plt.ylabel("population (in millions)")
plt.legend(["USA","China"])
plt.show()
```

```python
## Challenge #2: Compare the PERCENTAGE population growth of USA vs China

# Population of USA in the first available year
us_first_year_pop = us_data.population.iloc[0]
# Population of China in the first available year
china_first_year_pop = china_data.population.iloc[0]

plt.plot(us_data.year, us_data.population / us_first_year_pop * 100)
plt.plot(china_data.year, china_data.population / china_first_year_pop * 100)
plt.title("% population growth by country")
plt.xlabel("year")
plt.ylabel("% population growth (first year = 100)")
plt.legend(["USA","China"])
plt.show()
```
